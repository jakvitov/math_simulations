```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import math
```

#Simulace slabého zákona velkých čísel pro Gaussovo rozdělení
Zkusíme pro normální rozdělení otestovat větu, která říká, že při generování *n* pokusů, při stoupajícím *n*, součet rozdílů hodnot oproti střední hodnotě rozdělení konverguje k 0.



```python
mean = 4 #Střední hodnota pro dané rozdělení
std = 1   #Rozptyl
n = 100000 # Počet iterací
yvals = [] #Průběžné hodnoty konvergující k 0
#Vygenerujeme náhodnou posloupnost z Gaussova rozdělení
data = np.random.normal(mean, std, n)

#Sečteme rozdíly oproti střední hodnotě
sum_of_diffs = 0
for index, val in enumerate(data):
  sum_of_diffs += (val - mean)
  yvals.append(sum_of_diffs/(index + 1))

xvals = [i for i in range(n)]

plt.xlabel("Počet iterací")
plt.ylabel("Rozdíl od střední hodnoty")
plt.title("Konvergence sumy rozdílu od střední hodnoty pro Gaussovo rozdělení")
plt.plot(xvals, yvals)
plt.axhline(0, color='g')
plt.show()
```

# Simulace slabého zákona velkých čísel pro Poissonovo rozdělení
Zkusíme pro normální rozdělení otestovat větu, která říká, že při generování *n* pokusů, při stoupajícím *n*, součet rozdílů hodnot oproti střední hodnotě rozdělení konverguje k 0.


```python
plt.clf()
sum_of_diffs = 0
yvals = []
lmbd = 4 #Lambda parameter for the Poisson distribution

data = np.random.poisson(lmbd, n)

for index, val in enumerate(data):
  sum_of_diffs += (val - lmbd)
  yvals.append(sum_of_diffs/(index + 1))

plt.xlabel("Počet iterací")
plt.ylabel("Rozdíl od střední hodnoty")
plt.title("Konvergence sumy rozdílu od střední hodnoty pro Poissonovo rozdělení")
plt.plot(xvals, yvals)
plt.axhline(0, color='g')
plt.show()

```

# Simulace Lévyho-Lindbergovy věty pro Binomické rozdělení
V této simulaci se zaměřím na vizualizaci konvergence z vyjádření Centrální limitní věty - Lévyho Lindbergovy věty a to pro Binomické rozdělení. Tato věta je zobecněním Moivrovy-Laplaceovy věty a je platná pro rozdělení libovolného typu s konečnou střední hodnotu a konečným rozptylem.

```python
plt.clf()
bin_param_n = 10
bin_param_v = 0.1
log_mean = 15
n = 10000
bin_sums = []

#Data obsahující n stochasticky nezávislých hodnot binomického rozdělení


mean = bin_param_n * bin_param_v
diviation = mean * (1-bin_param_v)

for x in range(n):
  data_bin = np.random.binomial(bin_param_n, bin_param_v, n)
  #Přidáme hodnotu
  bin_sums.append((np.sum(data_bin) - (n * mean)) / math.sqrt(n) * math.sqrt(diviation))


distr = sns.kdeplot(bin_sums)
plt.xlabel("Hodnota součtu " + str(n) + " náhodných proměnných")
plt.ylabel("Četnost")
plt.title("Konvergence součtu náhodného vektoru k normálnímu rozdělení")
plt.show()



```

# Linear regression
Linear regression is the fundamental element of machine learning. Simply said, we try to fit a linear function between lineary correlated labels. Our goal is to chose weights $w_i$ and bias $b_i$ to minimalize the loss of a function $f(x_1,...,x_n) = (\sum_{i=1}^{n}(w_ix_i) + b)$. The whole problem of the loss and it's minimalization is elaborated in my [Gradient descend](https://github.com/jakvitov/math_simulations/blob/main/gradient_descend.ipynb) notebook.

## Try lin. regression on a data
Let's generate some linear data and try to fit a linear curve between them.

```python
import numpy as np
import matplotlib.pyplot as plt

x_features = np.linspace(-20, 30, 100)
y_labels = []

for x in np.nditer(x_features):
  y_labels.append(x * 5 + np.random.normal(-20, 40, size=1)[0] - 20)

y_labels = np.array(y_labels)

plt.figure()
plt.title("Randomly generated data with noise")
plt.xlabel("x")
plt.ylabel("$f(x)$")
plt.scatter(x_features, y_labels)
plt.show()
```

## Regression line
Now it is time to train a ML model to predict $f(x)$ based on the value of $x$.

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle

x_features, y_labels = shuffle(x_features, y_labels)

#Split for training and test
train_x = x_features[:80]
test_x = x_features[81:100]

train_y = y_labels[:80]
test_y = y_labels[81:100]
train_x = train_x.reshape((len(train_x), 1))

reg = linear_model.LinearRegression()
reg.fit(train_x, train_y)

pred_y = reg.predict(test_x.reshape((len(test_x), 1)))

print("Mean squared error: " + str(mean_squared_error(test_y, pred_y)))
print("Weight: " + str(reg.coef_))

fig = plt.figure()

plt.scatter(train_x, train_y, label="Training data")
plt.plot(test_x, pred_y, color="red", label="Predictions")
plt.title("Linear regression model on random data")
plt.xlabel("Input value")
plt.ylabel("Output value")
plt.grid()
plt.legend()
plt.show()



```
